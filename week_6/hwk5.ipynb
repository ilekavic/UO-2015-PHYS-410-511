```python
import numpy as np
import math

with open('C:/Users/IL/Documents/SciComp/HWK5/data.txt') as file:
    temp = [[float(digit) for digit in line.split()] for line in file]
print temp
```

Compute the average and standard deviation per decade

```python
nDecade = math.ceil(((temp[-1][0]-temp[0][0])/10))
print nDecade
ave = np.zeros(nDecade)
aveSqr = np.zeros(nDecade)
ctr = np.zeros(nDecade)

for i in range(len(temp)):
    index = math.floor((temp[i][0]-temp[0][0])/10)
    ave[index] += temp[i][1]
    aveSqr[index] += temp[i][1]**2
    ctr[index] += 1 
average = ave/ctr
averageSqr = aveSqr/ctr
print average
print averageSqr

std = np.sqrt(averageSqr - average**2)
print std
```

Formatted write commands into JSFiddle
    

```python
for i in range(int(nDecade)):
    testStr = ""
    for j in range(len(temp)):
        if(temp[j][0]> 1850 + 10*i and temp[j][0] < 1850 + 10*(i +1)):
           testStr += " ,"+str(temp[j][1])
    testStr += "],"
    print( str("[") + "%3d"  % (i))
    print(testStr)
    
   
```

Calculate slope in 30 year intervals

```python
slope = []
for i in range(13):
    slope.append((average[i+3]-average[i])/30.)
print slope
```

```python
for i in range(len(slope)):
    print( "['%3d-%3d' , %3.5f ]," % (1880 + 10*i , 1850+10*i,slope[i]))
```
