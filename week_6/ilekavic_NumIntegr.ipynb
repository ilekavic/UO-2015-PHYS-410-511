Random Sampling

```python
import numpy as np
import matplotlib.pyplot as plt

integral = []
error = []
iteration = 20
ave=0
for i in range(1,iteration +1,1):
    rand = np.random.random_sample(i)
    for j in range(i):
        ave += rand[j]**2
    integral.append(ave/float(i))
    error.append(1/3.-ave/float(i))
    ave = 0
print integral
x = np.arange(iteration)
plt.plot(x, error)
```

Midpoint Rule

```python

inter = 1.
val = 0
integral = []
error= []
for k in range(1,21,1):
    
    for i in range(k):
        val += ((i+.5)*(inter/k))**2
    integral.append(val*(inter/k))
    error.append(1/3.-val*(inter/k))
    val = 0
    
print integral
x = np.arange(20)
plt.plot(x, error)
```

Trapezoid Rule

```python

inter = 1.
val = 0
integral = []
error= []
for k in range(1,21,1):
    
    for i in range(k):
        val += (((i+1.)/k)**2+(i/float(k))**2)/2
    integral.append(val*inter/k)
    error.append(1/3.-val*inter/k)
    val = 0
    
print integral
x = np.arange(20)
plt.plot(x, error)
```

Simpson's Rule

```python
inter = 1.
val = 0
integral = []
error= []
for k in range(1,21,1):
    
    for i in range(k):
        val += (((i+1.)/k)**2+(i/float(k))**2+4*((2*i+1.)/(2*k))**2)/6.
    integral.append(val*inter/k)
    error.append(1/3.-val*inter/k)
    val = 0
    
print integral
x = np.arange(20)
plt.plot(x, error)
```
